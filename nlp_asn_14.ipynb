## STEP 3 — Import Libraries

```python

import spacy
from spacy import displacy
from transformers import pipeline, AutoTokenizer, AutoModelForTokenClassification
import torch
import pandas as pd
```


spaCy → Pretrained NLP pipeline  
Transformers → Contextual NER model  
torch → Backend for deep learning  
pandas → Data handling and tables  


## STEP 4 — Load spaCy Model

```python

nlp = spacy.load("en_core_web_sm")
```


spaCy pipeline includes:
- Tokenizer
- POS Tagger
- Dependency Parser
- Named Entity Recognizer


## STEP 5 — Sample Sentences

```python

sentences = [
    "Apple CEO Tim Cook announced new products in California.",
    "Elon Musk visited India to meet Prime Minister Narendra Modi.",
    "Microsoft acquired Activision Blizzard for $68 billion.",
    "The FIFA World Cup was held in Qatar in 2022.",
    "Google launched a new AI model in New York."
]
```

## STEP 6 — Apply spaCy NER

```python

spacy_results = []

for sentence in sentences:
    doc = nlp(sentence)
    for ent in doc.ents:
        print(f"Sentence: {sentence}")
        print(f"Entity: {ent.text}, Label: {ent.label_}")
        print("-"*40)
        spacy_results.append([sentence, ent.text, ent.label_])
```

## STEP 7 — Store spaCy Results in DataFrame

```python

spacy_df = pd.DataFrame(spacy_results, columns=["Sentence", "Entity", "Label"])
spacy_df
```

## STEP 8 — Visualize Entities (Optional)

```python

for sentence in sentences:
    doc = nlp(sentence)
    displacy.render(doc, style="ent", jupyter=True)
```

## STEP 9 — Load Hugging Face NER Model

```python

tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")
ner_pipeline = pipeline("ner", model=model, tokenizer=tokenizer)
```


Token classification assigns a label to each token.
Transformers use self-attention for contextual understanding.


## STEP 10 — Apply Transformer NER

```python

hf_results = []

for sentence in sentences:
    outputs = ner_pipeline(sentence)
    print(f"Sentence: {sentence}")
    for entity in outputs:
        print(entity)
        hf_results.append([sentence, entity["word"], entity["entity"], entity["score"]])
    print("-"*40)
```

## STEP 11 — Clean Transformer Output

```python

cleaned_results = []

for sentence, word, label, score in hf_results:
    clean_word = word.replace("##", "")
    cleaned_results.append([sentence, clean_word, label, score])

hf_df = pd.DataFrame(cleaned_results, columns=["Sentence", "Entity", "Label", "Confidence"])
hf_df
```


## STEP 12 — Comparison Table

| Feature | spaCy | Hugging Face |
|----------|--------|--------------|
| Model Type | Statistical | Transformer |
| Speed | Fast | Slower |
| Context Handling | Limited | Strong |
| Confidence Score | No | Yes |
| Accuracy | Moderate | High |



## STEP 13 — Result Analysis

Transformers detected more contextual entities and provided confidence scores.
